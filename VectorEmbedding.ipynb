```python
import gensim.downloader as api
model = api.load("word2vec-google-news-300")
```

```python
word_vector = model
print(word_vector['Man'])
print(word_vector['cat'].shape)
```

```python
print(word_vector.most_similar(positive=['King','Woman'], negative=['Man'],topn=10))
```

```python
print(word_vector.most_similar('tower',topn=5))
```

```python
print(word_vector.similarity('man','woman'))
print(word_vector.similarity('man','king'))
print(word_vector.similarity('rock','paper'))
print(word_vector.similarity('fool','woman'))
```

## creating vector embedding 

```python
import torch
input_ids = torch.tensor([1,3,5,2])

vocab_size =6
output_dim = 3
torch.manual_seed(123)
embedding_layer=  torch.nn.Embedding(vocab_size,output_dim)
```

```python
print(embedding_layer.weight)
```

```python
print(embedding_layer(torch.tensor([3])))
```

```python
print(embedding_layer(input_ids))
```
